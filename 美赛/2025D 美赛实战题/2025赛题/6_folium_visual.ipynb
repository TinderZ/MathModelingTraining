# 通过folium可视化地图

```python
import folium
import pandas as pd
# 读取CSV文件
df = pd.read_csv('top_10_nodes.csv')
```

```python
#创建一个地图对象，初始位置为选中节点的中心点
cen_lat = 39.293769
cen_lon = -76.612189
map1 = folium.Map(location=[cen_lat, cen_lon], zoom_start=15)

# 遍历每一行，添加标记
for index, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"OSM ID: {row['osmid']}"
    ).add_to(map1)

# 保存地图为HTML文件
map1.save('map1.html')
map1
```
